```python
# Now what I want is to extract Top 5 countries in each year.

# loop each year to get top 5 countries 

top_5_happy = pd.DataFrame()

# Loop through each unique year in the DataFrame
for col_year in df["Year"].unique():
    # Filter the DataFrame for the current year
    top_5_current = df[df["Year"] == col_year]
    
    # Get the top 5 rows by 'Economy (GDP per Capita)'
    top_5_current = top_5_current.nlargest(5, "Happiness Score")  
    
    # Optionally, sort the top 5 by 'Economy (GDP per Capita)'
    top_5_current = top_5_current.sort_values("Happiness Score",ascending=False)
    
    # Append the results to the main DataFrame
    top_5_happy = pd.concat([top_5_happy, top_5_current])

# Reset the index of the final DataFrame for cleaner output
top_5_happy.reset_index(drop=True, inplace=True)

# Display the result
top_5_happy
```

```python
# Now what I want is to extract Top 5 countries in each year.

# loop each year to get top 5 countries 
low_5_happy = pd.DataFrame()


# Loop through each unique year in the DataFrame
for col_year in df["Year"].unique():
    # Filter the DataFrame for the current year
    low_5_current = df[df["Year"] == col_year]
    
    # Get the top 5 rows by 'Economy (GDP per Capita)'
    low_5_current = low_5_current.nsmallest(5, "Happiness Score")  
    
    # Optionally, sort the top 5 by 'Economy (GDP per Capita)'
    low_5_current = low_5_current.sort_values("Happiness Score",ascending=False)
    
    # Append the results to the main DataFrame
    low_5_happy = pd.concat([low_5_happy, low_5_current])

# Reset the index of the final DataFrame for cleaner output
low_5_happy.reset_index(drop=True, inplace=True)

# Display the result
low_5_happy
```

```python

# loop each year
for year in Switzerland_df["Year"]:
    Switzerland_df.loc[Switzerland_df["Year"] == year,"cor_Economy"] =Switzerland_df.loc[Switzerland_df["Year"] == year,"Happiness Score"]/Switzerland_df.loc[Switzerland_df["Year"] == year,"Economy (GDP per Capita)"]

for year in Switzerland_df["Year"]:
    Switzerland_df.loc[Switzerland_df["Year"] == year,"cor_Health"] =Switzerland_df.loc[Switzerland_df["Year"] == year,"Happiness Score"]/Switzerland_df.loc[Switzerland_df["Year"] == year,"Health (Life Expectancy)"]

for year in Switzerland_df["Year"]:
    Switzerland_df.loc[Switzerland_df["Year"] == year,"cor_Freedom"] =Switzerland_df.loc[Switzerland_df["Year"] == year,"Happiness Score"]/Switzerland_df.loc[Switzerland_df["Year"] == year,"Freedom (life choices)"]

for year in Switzerland_df["Year"]:
    Switzerland_df.loc[Switzerland_df["Year"] == year,"cor_Generosity"] =Switzerland_df.loc[Switzerland_df["Year"] == year,"Happiness Score"]/Switzerland_df.loc[Switzerland_df["Year"] == year,"Generosity"]

for year in Switzerland_df["Year"]:
    Switzerland_df.loc[Switzerland_df["Year"] == year,"cor_Trust"] =Switzerland_df.loc[Switzerland_df["Year"] == year,"Happiness Score"]/Switzerland_df.loc[Switzerland_df["Year"] == year,"Trust (Government Corruption)"]



Switzerland_df
```
